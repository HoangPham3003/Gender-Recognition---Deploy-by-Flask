```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
import re
```

```python
def getNew(news):
    title = ''
    description = ''
    publish_time = ''
    link = ''
    
    content = news.find('div', 'news-item__content')
    
    linkStr = content.find('a')
    title = linkStr.text.strip() # Title
    link = linkStr['href'] # Link
    description = content.find('a', 'news-item__sapo').text.strip()
    description = description.replace('(Dân trí)', '').strip()
    description = description.replace('-', '').strip()
    publish_time = content.find('span', 'news-item__time').text.strip()
    return {'Title': title, 'Description': description, 'Publish_Time': publish_time, 'Link': link}
```

```python
url = "https://dantri.com.vn/su-kien.htm"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
news = soup.find_all('div', 'news-item news-item--timeline news-item--left2right')
data = []
for new in news:
    row = getNew(new)
    data.append(row)

df = pd.DataFrame(data)
df.head(10)
```

```python
df.to_csv('articles.csv')
```

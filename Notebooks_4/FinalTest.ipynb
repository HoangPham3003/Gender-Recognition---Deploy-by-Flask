#### Remove customerID and Churn column

```python
myData = rawData.copy()
myData = myData.drop(columns=['customerID', 'Churn'], axis=1)
myData.head(5)
```

#### Encode category values

```python
myData = myData.apply(preprocessing.LabelEncoder().fit_transform)
myData.head(5)
```

#### Modeling

```python
X = myData.values
X[0:10]
```

#### Normalization

```python
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:10]
```

#### Choosing the number of cluster

```python
# Choosing number of cluster using Within-Cluster-Sum-of-Squares (WCSS) and Elbow method
wcss = []
for i in range(1, 12):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 8))
plt.plot(range(1, 12), wcss)
plt.title('WCSS by Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### ==> Choose 3 cluster

#### K-Means

```python
# Create model
clusterNum = 3
kmeans = KMeans(init='k-means++', n_clusters=clusterNum)
kmeans.fit(X)
labels = kmeans.labels_
labels[0:20]
```

#### Add label to rawData

```python
rawData['label'] = labels
rawData.head(5)
```

#### Data preparation

```python
X = myDf.drop('label', axis=1).values
X[0:10]
```

```python
y = myDf['label'].values
y[0:10]
```

#### Normalize data

```python
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:10]
```

#### Split train-test data set

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=3)
print("Shape of train data set: ", train_x.shape, train_y.shape)
print("Shape of test data set: ", test_x.shape, test_y.shape)
```

#### Modeling using SVM

```python
from sklearn import metrics
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix
import itertools
```

```python
mySVM = svm.SVC(kernel='rbf')
mySVM.fit(train_x, train_y)

# Predict
predictSVM = mySVM.predict(test_x)
print(predictSVM[0:20])
print(test_y[0:20])
print()

# Evaluation
trainACS_SVM = metrics.accuracy_score(train_y,  mySVM.predict(train_x))
testACS_SVM =  metrics.accuracy_score(test_y,  predictSVM)

print("SVM Train data set's Accuracy: ", trainACS_SVM)
print("SVM Test data set's Accuracy: ", testACS_SVM)
```

#### Using confusion matrix to evaluate model

```python
# Confusion matrix function
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(test_y, predictSVM, labels=[0, 1, 2])
np.set_printoptions(precision=2)
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['0', '1', '2'], normalize=True,  title='Confusion matrix')
```

```python
# Import libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Read data from csv file
data = pd.read_csv(r"ApartmentPrice.csv")
```

```python
print(data.head(10))              # Show all data
```

```python
print(data.describe())            # Sumarize data
```

```python
print(data.keys())                # Data attributes
```

```python
# Select attributes
# Attribute description
# SalePrice: price (USD)
# YearBuilt
# Size: size(sqf)
# Floor: int
# N_FacilitiesNearBy(Total): mall, department store, hospital,...
# N_SchoolNearBy(Total): elementary, middle, high, university
df = data[['SalePrice', 'YearBuilt', 'Size', 'Floor', 'N_FacilitiesNearBy(Total)', 'N_SchoolNearBy(Total)']]
```

```python
# Exploratory data analysis
print(df.head(10))
```

```python
# Distributiion of each feature
# SalePrice
plt.hist(df['SalePrice'])
plt.xlabel('SalePrice')
plt.show()
# ==============================================
# YearBuilt
plt.hist(df['YearBuilt'])
plt.xlabel('YearBuilt')
plt.show()
# ==============================================
# Size
plt.hist(df['Size'])
plt.xlabel('Size')
plt.show()
# ==============================================
# Floor
plt.hist(df['Floor'])
plt.xlabel('Floor')
plt.show()
# ==============================================
# N_FacilitiesNearBy(Total)
plt.hist(df['N_FacilitiesNearBy(Total)'])
plt.xlabel('N_FacilitiesNearBy(Total)')
plt.show()
# ==============================================
# N_SchoolNearBy(Total)
plt.hist(df['N_SchoolNearBy(Total)'])
plt.xlabel('N_SchoolNearBy(Total)')
plt.show()
```

```python
# Relationship between SalePrice and other features
```

```python
# SalePrice and YearBuilt
plt.scatter(df['YearBuilt'], df['SalePrice'], color='blue')
plt.xlabel('YearBuilt')
plt.ylabel('SalePrice')
plt.title('Relationship between YearBuilt and SalePrice')
plt.show()
# ==============================================
# SalePrice and Size
plt.scatter(df['Size'], df['SalePrice'], color='blue')
plt.xlabel('Size')
plt.ylabel('SalePrice')
plt.title('Relationship between Size and SalePrice')
plt.show()
# ==============================================
# SalePrice and Floor
plt.scatter(df['Floor'], df['SalePrice'], color='blue')
plt.xlabel('Floor')
plt.ylabel('SalePrice')
plt.title('Relationship between Floor and SalePrice')
plt.show()
# ==============================================
# SalePrice and N_FacilitiesNearBy(Total)
plt.scatter(df['N_FacilitiesNearBy(Total)'], df['SalePrice'], color='blue')
plt.xlabel('N_FacilitiesNearBy(Total)')
plt.ylabel('SalePrice')
plt.title('Relationship between N_FacilitiesNearBy(Total) and SalePrice')
plt.show()
# ==============================================
# SalePrice and N_SchoolNearBy(Total)
plt.scatter(df['N_SchoolNearBy(Total)'], df['SalePrice'], color='blue')
plt.xlabel('N_SchoolNearBy(Total)')
plt.ylabel('SalePrice')
plt.title('Relationship between N_SchoolNearBy(Total) and SalePrice')
plt.show()
```

```python
# ==============================================================================
# Modeling
# Input data: SalePrice and Size
# Predict model: Linear regression
# ==============================================================================
X = np.asanyarray(df[['Size']])
Y = np.asanyarray(df[['SalePrice']])

# Split data: 80% for training and 20% for testing
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size=0.2)
# Training data set distribution
plt.scatter(train_x, train_y, color='blue')
plt.xlabel('Size')
plt.ylabel('SalePrice')
plt.title('Training data set')
plt.show()

# Testing data set distribution
plt.scatter(test_x, test_y, color='blue')
plt.xlabel('Size')
plt.ylabel('SalePrice')
plt.title('Testing data set')
plt.show()

# Modeling -> Equation
model_1 = LinearRegression()
model_1.fit(train_x, train_y)
print("Coefficient: ", model_1.coef_)
print("Intercept: ", model_1.intercept_)

coefs = model_1.coef_
intercept = model_1.intercept_

theta_0 = intercept[0]
theta_1 = coefs[0][0]

print("Equation: y = {:.2f} + {:.2f}x".format(theta_0, theta_1))

# Plot fit line
plt.scatter(train_x, train_y, color='blue')
plt.xlabel('Size')
plt.ylabel('SalePrice')
plt.plot(train_x, theta_0+theta_1*train_x, color='red')
plt.title('Training data fit in a linear regression')
plt.show()

# Evaluation ( R2 score )
# Training data set
print("R2-score (Training data set): %.2f" % r2_score(train_y, model_1.predict(train_x)))
# Testing data set
print("R2-score (Testing data set): %.2f" % r2_score(test_y, model_1.predict(test_x)))
```

```python
# ==============================================================================
# Modeling
# Input data: SalePrice and Size
# Predict model: Polynomial regression with degree = 2
# ==============================================================================
X = np.asanyarray(df[['Size']])
Y = np.asanyarray(df[['SalePrice']])

# Split data: 80% for training and 20% for testing
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size=0.2)
# Training data set distribution
plt.scatter(train_x, train_y, color='blue')
plt.xlabel('Size')
plt.ylabel('SalePrice')
plt.title('Training data set')
plt.show()

# Testing data set distribution
plt.scatter(test_x, test_y, color='blue')
plt.xlabel('Size')
plt.ylabel('SalePrice')
plt.title('Testing data set')
plt.show()

# Polinomial Transform
poly = PolynomialFeatures(degree=2)
train_x_poly = poly.fit_transform(train_x)
# print(train_x_poly)

# Modeling -> equation
model_2 = LinearRegression()
model_2.fit(train_x_poly, train_y)

print("Coefficient: ", model_2.coef_)
print("Intercept: ", model_2.intercept_)

coef_0, coef_1, coef_2 = model_2.coef_[0]
intercept = model_2.intercept_[0]

print("Equation: y = {:.2f} + {:.2f}x + {:.2f}x^2".format(intercept, coef_1, coef_2))

# Plot fit line
xx = np.arange(0.0, float(max(train_x)), 0.1)
yy = intercept + coef_1*xx + coef_2*(xx**2)

plt.scatter(train_x, train_y, color='blue')
plt.xlabel('Size')
plt.ylabel('SalePrice')
plt.plot(xx, yy, color='red')
plt.title('Training data fit in a polynomial regression')
plt.show()

# Evaluation ( R2 score )
# Training data set
print("R2-score (Training data set): %.2f" % r2_score(train_y, model_2.predict(train_x_poly)))
# Testing data set
test_x_poly = poly.fit_transform(test_x)
print("R2-score (Testing data set): %.2f" % r2_score(test_y, model_2.predict(test_x_poly)))
```

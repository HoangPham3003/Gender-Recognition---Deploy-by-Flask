- Language: Moderate use of profanity.
- Lyrics: References to profanity, sexuality, violence, alcohol, or drug use in music.
- Mature_Humor: Depictions or dialogue involving "adult" humor, including sexual references.

- Mild_Lyrics: Mild References to profanity, sexuality, violence, alcohol, or drug use in music.
- MildSuggestiveThemes: some provocative references or materials
- Mild_Violence: Some scenes involving aggressive conflict.
- No_Descriptors: No content descriptors.
- Nudity: Graphic or prolonged depictions of nudity.
- Partial_Nudity: Brief and/or mild depictions of nudity.
- Sexual_Content: Non-explicit depictions of sexual behavior, possibly including partial nudity.
- Sexual_Themes: References to sex or sexuality.
- Simulated_Gambling: Player can gamble without betting or wagering real cash or currency.

- Suggestive_Themes: Provocative references or materials.

- UseofDrugsandAlcohol: The consumption of alcoholic and drugs beverages.
- Violence: Scenes involving aggressive conflict. May contain bloodless dismemberment.
- ESRB_rating: rating: RP - EC - E - E+10 - T - M - A

<span style="font-weight:bold">b) Value</span>
- console: 
    + 0 = PS4
    + 1 = PS4 & Xbox_one
- ESRB_rating:
    + RP: Rating Pending (None rating)
    + EC: Early Childhood
    + E: Everyone

<h2>5. Games-Features Data Preparation</h2>

#### Evaluate relationship between 'esrb_rating' and other features

```python
# Encode values of esrb_rating key
le = preprocessing.LabelEncoder()
gamesInfo['digit_rating'] = le.fit(gamesInfo['esrb_rating'].unique()).transform(gamesInfo['esrb_rating'])
gamesInfo.head(5)
```

```python
# Correlation matrix
corr = gamesInfo.corr()
plt.figure(figsize=(15, 15))
sn.heatmap(corr)
```

```python
corr_rating = corr[['digit_rating']].copy(deep=True)
corr_rating
```

```python
# Correlation between 'digit_rating' and other features
corr_rating = corr_rating.drop(index=['digit_rating'], axis=0, inplace=False)
corr_rating[['digit_rating']] = abs(corr_rating[['digit_rating']].values)
corr_rating = corr_rating[['digit_rating']].sort_values(by=['digit_rating'], axis=0, ascending=False, inplace=False)

# Bar plot
corr_values, corr_indexs = corr_rating['digit_rating'].values, corr_rating.index
plt.figure(figsize=(15,15))
ax = sn.barplot(x=corr_values, y=corr_indexs)
for i, v in enumerate(corr_values):
    ax.text(v, i, str(v), color='blue', fontweight='bold')
plt.show()
```

#### Select important features

```python
TrueKeys = [x for x in corr_rating[['digit_rating']].index if corr_rating[['digit_rating']].loc[x].values[0] >= 0.1]
TrueKeys.insert(0, 'title')
TrueKeys.append('esrb_rating')
TrueKeys
```

#### Create games-features data

```python
games_features = gamesInfo.copy()
games_features = games_features.drop(columns=[x for x in gamesInfo.keys() if x not in TrueKeys], axis=1)
games_features.head(5)
```

```python
games_features.shape
```

#### Adding features related to age

```python
for index, row in games_features.iterrows():
    x = row['esrb_rating'] 
    if x == 'ET':
        games_features.at[index, '10+'] = 1.
    elif x == 'T':
        games_features.at[index, '13+'] = 1.
    elif x == 'M':
        games_features.at[index, '17+'] = 1.
    elif x == 'E':
        games_features.at[index, '10+'] = 1.
        games_features.at[index, '13+'] = 1.
        games_features.at[index, '17+'] = 1.

games_features = games_features.fillna(0)
games_features = games_features.drop(['esrb_rating'], axis=1)
games_features = games_features.astype(dtype={'10+': 'int64', '13+': 'int64', '17+': 'int64'})
games_features.head(10)
```

```python
games_features.info()
```

<h2>6. User Data Preparation</h2>

#### Create user rating input

```python
userRatingInput = [
            {'title': 'Mugsters', 'rating': 5},
            {'title': 'Unearthing Mars2', 'rating': 1},
            {'title': 'Dungeons 3', 'rating': 2},
            {'title': 'The Sims™ 4 Deluxe Party Pack', 'rating': 5},
            {'title': 'Citadel: Forged with Fire', 'rating': 4.5},
            {'title': 'WARHAMMER 40,000 : DEATHWATCH', 'rating': 4.5},
            {'title': 'The Walker', 'rating': 3},
            {'title': 'One Piece: Grand Cruise', 'rating': 4},
            {'title': 'Psychonauts', 'rating': 3.5},
            {'title': 'The Legend of Heroes: Trails of Cold Steel III', 'rating': 5}
         ] 
df_userRatingInput = pd.DataFrame(data=userRatingInput)
df_userRatingInput
```

#### Add games-features to the user input

```python
games_users = games_features[games_features['title'].isin(df_userRatingInput['title'].tolist())]
games_users = pd.merge(games_users, df_userRatingInput)
games_users
```

#### Create user-rating data

```python
user_ratings = games_users['rating']
user_ratings
```

#### Drop columns we won't use

```python
games_users = games_users.drop(columns=['title', 'rating'], axis=1)
games_users
```

<h2>7. Content-based Recommender System</h2>

#### Calculate user profile rating

```python
userProfileRating = games_users.transpose().dot(user_ratings)
userProfileRating
```

#### Create weighted games table

```python
weightedGames = games_features.copy()
# Reset index belong to games's title
weightedGames = weightedGames.set_index(weightedGames['title'])
# Drop games that were played
weightedGames = weightedGames.drop(index=df_userRatingInput['title'], axis=0)
# Drop features we won' use for caculating
weightedGames = weightedGames.drop(columns=['title'], axis=1)
weightedGames
```

#### Calculate recommendation table

```python
recommendationTable = ((weightedGames*userProfileRating).sum(axis=1))/(userProfileRating.sum())
recommendationTable = recommendationTable.sort_values(ascending=False)
recommendationTable.head(10)
```

```python
df_recommendationTable = gamesInfo.loc[gamesInfo['title'].isin(recommendationTable.head(10).index)]
df_recommendationTable
```
